```python
x = np.linspace(-5, 5, 20)

plt.plot(x, np.sin(x), 'g-', label='sin(x)')
plt.plot(x, np.cos(x), 'r--', label='cos(x)')

plt.legend()
plt.show()
```

```python
#BAR PLOT
fig, ax = plt.subplots()
languages = ['English', 'Spanish', 'French', 'German', 'Italian']
people = [100, 200, 300, 400, 500]
ax.bar(languages, people, color='skyblue')  # You can specify a color if needed

plt.xlabel('Languages')
plt.ylabel('People')
plt.title('Number of People by Language')
plt.show()
```

```python
#PIE CHART
fig1 = plt.figure()
ax1 = fig1.add_axes([0, 0, 1, 1])
languages = ['English', 'Spanish', 'French', 'German', 'Italian']
people = [100, 200, 300, 400, 500]
ax.pie(people, labels = languages, autopct='%1.2f%%')
plt.show()
```

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x, y,color='g')
ax.plot(x, z, 'r--', label='cos(x)')
ax.legend()
plt.show()
```

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20* np.random.rand(1000)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x, y, c=z, cmap='viridis', s=100)
```
