```python
v1 = np.array([0,0,2,3])
v2 = np.array([0,0,-3,3])

sum = v1 + v1
print(sum)

diff = v2-v1
print(diff)

mul = 2*v1*v2
print(mul)
```

```python
#DOT PRODUCT
a = np.array([2,1,4])
b = np.array([2,4,5])
c = np.dot(a, b)

print(c)
```

```python
#CROSS PRODUCT

a = np.array([2,1,4])
b = np.array([2,4,5])
c = np.cross(a, b)

print(c)
```

```python
import numpy as np

a = np.array([2, 1, 4])
v = np.array([2, 4, 5])

# Projection using dot product
projection = (np.dot(a, v) / np.linalg.norm(v)**2) * v
print(projection)
```
